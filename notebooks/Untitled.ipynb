```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(4)+1
y = 6.0+2.0*np.random.rand(4)
plt.bar(x,y)
plt.ylim([6,8])
plt.grid()
plt.xlabel('x')
plt.xticks(x)
plt.show()
```

```python
x = np.linspace(0,2,1000)
y = np.exp(-x)
y1 = np.exp(-2.0*x)

plt.semilogy(x,y,label='$e^{-x}$')
plt.semilogy(x,y1,label='$e^{-2x}$')
plt.legend()
plt.grid(which='both')
plt.savefig('plot.jpg')
```

```python
hour,temperature,precip_prob = np.loadtxt('../data/weather.csv',dtype=np.float,delimiter=',',skiprows=1,unpack=True)

plt.subplot(1,2,1)
plt.plot(hour,temperature)
plt.xlabel('Hour')
plt.ylabel('Temperature ($^o$F)')
plt.title('Temperature')

plt.subplot(1,2,2)
plt.bar(hour,precip_prob)
plt.xlabel('Hour')
plt.ylabel('Probability (%)')
plt.title('Probability of Precipitation')
plt.show()
```
